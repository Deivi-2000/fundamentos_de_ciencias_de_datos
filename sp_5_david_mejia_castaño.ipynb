## <span style="color:#2F749F;"><strong>Ejercicio 1. Limpieza e imputación de un conjunto de datos mixto</strong></span>

1. Crea el DataFrame base **df** ejecutando el siguiente código:
    ```python
    df = pd.DataFrame({
        'Nombre': ['Ana', 'Luis', 'Pedro', None, 'Marta', 'Luis', 'Sofía'],
        'Edad': [25, np.nan, 35, 29, -5, 25, None],
        'Ciudad': ['Bogotá', 'Medellín', None, 'Medellín', 'Cali', 'Bogotá', 'Cali'],
        'Ingreso': [3500, 4800, np.nan, 5200, 5100, np.nan, 4700],
        'FechaIngreso': ['2023-01-01', '2023-01-05', None, '2023-01-10', '2023-01-12', None, '2023-01-15']
    })

2. En un nuevo notebook, aplica las etapas del proceso de limpieza e imputación de datos que consideres necesarias sobre el DataFrame df. Ten encuenta incluir:
    - Código correctamente comentado y ejecutado.
    - Explicaciones breves de cada paso.
    - Resultados visuales o estadísticos que evidencien las transformaciones.

````python
import pandas as pd
import numpy as np
````

````python
#Creamos el Dataframe

df = pd.DataFrame({
    'Nombre': ['Ana', 'Luis', 'Pedro', None, 'Marta', 'Luis', 'Sofía'],
    'Edad': [25, np.nan, 35, 29, -5, 25, None],
    'Ciudad': ['Bogotá', 'Medellín', None, 'Medellín', 'Cali', 'Bogotá', 'Cali'],
    'Ingreso': [3500, 4800, np.nan, 5200, 5100, np.nan, 4700],
    'FechaIngreso': ['2023-01-01', '2023-01-05', None, '2023-01-10',
                     '2023-01-12', None, '2023-01-15']
})

print("DataFrame original")
print(df)
print("\nTipos de datos")
print(df.dtypes)
print("\nValores nulos por columna")
print(df.isnull().sum())
````

Acá se identifican los valores faltantes y los tipos de variables para escoger la imputación adecuada.

````python
# Copia para limpieza
df_limpio = df.copy()

# Convertir fecha a formato datetime
df_limpio['FechaIngreso'] = pd.to_datetime(df_limpio['FechaIngreso'], errors='coerce')

# Corregir edad inválida: las edades negativas se consideran faltantes
df_limpio['Edad'] = df_limpio['Edad'].apply(lambda x: np.nan if pd.notnull(x) and x < 0 else x)

print("\n Después de corregir edad negativa y convertir fechas")
print(df_limpio)

print("\n Valores faltantes después de la corrección inicial")
print(df_limpio.isnull().sum())
````

La edad `-5` no es válida, por eso se transforma en `NaN`. Las fechas se convierten con `pd.to_datetime()`.

````python
# Imputación de variables
df_limpio['Nombre'] = df_limpio['Nombre'].fillna('Desconocido')
df_limpio['Edad'] = df_limpio['Edad'].fillna(df_limpio['Edad'].median())
df_limpio['Ciudad'] = df_limpio['Ciudad'].fillna(df_limpio['Ciudad'].mode()[0])
df_limpio['Ingreso'] = df_limpio['Ingreso'].fillna(df_limpio['Ingreso'].median())
df_limpio['FechaIngreso'] = df_limpio['FechaIngreso'].ffill()

print("\n DataFrame limpio e imputado")
print(df_limpio)

print("\n Valores faltantes finales")
print(df_limpio.isnull().sum())

print("\n Estadísticas después de imputar")
print(df_limpio.describe())
````

Se usa mediana para variables numéricas, moda para ciudad, categoría `Desconocido` para nombre y `ffill()` para completar fechas de forma secuencial.


````python
#Realizamos una comparación estadística
print("Comparación Estadística: Antes vs Después")

print("\n Valores nulos por columna:")
comparacion_nulos = pd.DataFrame({
    'Antes': df.isnull().sum(),
    'Después': df_limpio.isnull().sum()
})
print(comparacion_nulos)

print("\n Estadísticas numéricas (Edad e Ingreso):")
for col in ['Edad', 'Ingreso']:
    print(f"\n  {col}:")
    print(f"    Antes  → Media: {df[col].mean():.1f} | Mediana: {df[col].median():.1f} | Nulos: {df[col].isnull().sum()}")
    print(f"    Después→ Media: {df_limpio[col].mean():.1f} | Mediana: {df_limpio[col].median():.1f} | Nulos: {df_limpio[col].isnull().sum()}")
````

## <span style="color:#2F749F;"><strong>Ejercicio 2. Limpieza de duplicados</strong></span>

1. Crea el DataFrame base **df** ejecutando el siguiente código:
    ```python
    df = pd.DataFrame({
        'ID': [101, 102, 102, 103, 104, 104, 104],
        'Nombre': ['Ana', 'Luis', 'Luis', 'Marta', 'Carlos', 'Carlos', 'Carlos'],
        'Edad': [25, 30, 30, 29, 40, 40, 41],
        'Ciudad': ['Bogotá', 'Cali', 'Cali', 'Medellín', 'Cali', 'Cali', 'Cali'],
        'FechaRegistro': ['2023-01-01', '2023-01-05', '2023-01-05', '2023-01-10',
                        '2023-01-15', '2023-01-15', '2023-01-16']
    })

2.  En el mismo notebook anterior, para el nuevo dataframe df, responde a las siguientes preguntas (utilizando python):
    - ¿Cuál es el total de registros originales?
    - ¿Cuáles y cuántos son los duplicados exactos?
    - ¿Cuáles y cuántos son los duplicados por varias columnas?
    - ¿Cuántos registros debes eliminar?
    - ¿Cuántos registros quedan después de la limpieza?

````python
#Creamos el Dataframe

df = pd.DataFrame({
    'ID': [101, 102, 102, 103, 104, 104, 104],
    'Nombre': ['Ana', 'Luis', 'Luis', 'Marta', 'Carlos', 'Carlos', 'Carlos'],
    'Edad': [25, 30, 30, 29, 40, 40, 41],
    'Ciudad': ['Bogotá', 'Cali', 'Cali', 'Medellín', 'Cali', 'Cali', 'Cali'],
    'FechaRegistro': ['2023-01-01', '2023-01-05', '2023-01-05', '2023-01-10',
                      '2023-01-15', '2023-01-15', '2023-01-16']
})

print("DataFrame original")
print(df)

total_original = len(df)
print("\n Total de registros:", total_original)
````

Se identifican duplicados exactos y duplicados por columnas

````python
# Duplicados exactos
duplicados_exactos = df[df.duplicated(keep=False)]
exactos_a_eliminar = df.duplicated().sum()

print("Duplicados exactos")
print(duplicados_exactos)

print("\n Filas involucradas en duplicados exactos:", len(duplicados_exactos))
print("\n Registros exactos a eliminar:", exactos_a_eliminar)
````

````python
# Duplicados por varias columnas
columnas = ['ID', 'Nombre', 'Ciudad']
duplicados = df[df.duplicated(subset=columnas, keep=False)]

print("Duplicados por columnas clave:", columnas)
print(duplicados)

print("\n Filas involucradas en duplicados por columnas clave:", len(duplicados))
````

````python
# Limpieza final: se conserva el primer registro de cada grupo según las columnas clave
df_sin_duplicados = df.drop_duplicates(subset=columnas, keep='first')

registros_eliminados = total_original - len(df_sin_duplicados)

print("Registros que se deben eliminar:", registros_eliminados)
print("\n Registros después de la limpieza:", len(df_sin_duplicados))
````

Hay duplicados exactos y también registros repetidos por ID, Nombre y Ciudad. Al conservar el primer registro por grupo quedan 4 registros.

## <span style="color:#2F749F;"><strong>Ejercicio 3. Corrección de errores tipográficos o de codificación</strong></span>


1. Crea el DataFrame base **df** ejecutando el siguiente código:
    ```python
    df = pd.DataFrame({
        'Ciudad': ['bogota', 'Bogotá', 'BOGOTA', 'bogotá', 'bogata', 'Bógota', 'BogoTa', 'Cali', 'calí', 'medellín', 'medellin']
    })

2.  En el mismo notebook anterior, para el nuevo dataframe df, estandariza la variable ciudad utilizando python.

````python
import unicodedata

df = pd.DataFrame({
    'Ciudad': ['bogota', 'Bogotá', 'BOGOTA', 'bogotá', 'bogata', 'Bógota', 'BogoTa', 'Cali', 'calí', 'medellín', 'medellin']
})

def quitar_tildes(texto):
    texto = str(texto).lower().strip()
    texto = ''.join(
        c for c in unicodedata.normalize('NFD', texto)
        if unicodedata.category(c) != 'Mn'
    )
    return texto

# Normalización básica
df['Ciudad_limpia'] = df['Ciudad'].apply(quitar_tildes)

# Diccionario de reemplazo
reemplazos = {
    'bogota': 'Bogotá',
    'bogata': 'Bogotá',
    'cali': 'Cali',
    'medellin': 'Medellín'
}

df['Ciudad_estandarizada'] = df['Ciudad_limpia'].replace(reemplazos)

print("\n Ciudades estandarizadas")
print(df)

print("\n Conteo final")
print(df['Ciudad_estandarizada'].value_counts())
````

## <span style="color:#2F749F;"><strong>Ejercicio 5. Imputación con KNN y regresión</strong></span>

1. Crea el DataFrame base ejecutando el siguiente código:

    ```python
        import pandas as pd
        import numpy as np

        np.random.seed(42)

        n = 30

        df = pd.DataFrame({
            'Edad': np.random.randint(18, 60, n),
            'Promedio': np.round(np.random.uniform(2.0, 5.0, n), 2),
            'Horas_estudio': np.random.randint(0, 40, n),
            'Ingreso_familiar': np.random.randint(800000, 8000000, n),
            'Semestres_cursados': np.random.randint(1, 10, n)
        })

        # Introducción de valores faltantes con distintos patrones
        df.loc[[2, 5, 8, 15], 'Promedio'] = np.nan
        df.loc[[1, 7, 10, 18], 'Horas_estudio'] = np.nan
        df.loc[[3, 6, 14], 'Ingreso_familiar'] = np.nan

2. Al analizar el dataframe `df`:
   - ¿Qué variables tienen mayor proporción de datos faltantes?
   - ¿Observas algún patrón estructurado o parecen aleatorios?

3. Realiza la imputación de valores faltantes utilizando KNN y regresión
4. Analiza si se preserva la naturaleza de los datos con cada una de las técnicas de imputación (KNN y regresión)

````python
from sklearn.impute import KNNImputer
from sklearn.linear_model import LinearRegression
````

````python
# DataFrame
np.random.seed(42)

n = 30

df = pd.DataFrame({
    'Edad': np.random.randint(18, 60, n),
    'Promedio': np.round(np.random.uniform(2.0, 5.0, n), 2),
    'Horas_estudio': np.random.randint(0, 40, n),
    'Ingreso_familiar': np.random.randint(800000, 8000000, n),
    'Semestres_cursados': np.random.randint(1, 10, n)
})

# Introducción de valores faltantes con distintos patrones
df.loc[[2, 5, 8, 15], 'Promedio'] = np.nan
df.loc[[1, 7, 10, 18], 'Horas_estudio'] = np.nan
df.loc[[3, 6, 14], 'Ingreso_familiar'] = np.nan
````

````python
#Proporción de datos faltantes

print("DataFrame con faltantes")
print(df)

faltantes = pd.DataFrame({
    'faltantes': df.isnull().sum(),
    'porcentaje': (df.isnull().mean() * 100).round(2)
})

print("\n Proporción de datos faltantes")
print(faltantes)
````

Las variables con mayor proporción de faltantes son Promedio y Horas_estudio, ambas con 13.33%. No se observa un patrón estructurado claro, pues parecen faltantes artificiales ubicados en índices específicos.

````python
# Imputación con KNN
imputer = KNNImputer(n_neighbors=3)

df_knn = pd.DataFrame(
    imputer.fit_transform(df),
    columns=df.columns
)

# Ajuste de naturaleza de los datos
df_knn['Edad'] = df_knn['Edad'].round().astype(int)
df_knn['Promedio'] = df_knn['Promedio'].round(2)
df_knn['Horas_estudio'] = df_knn['Horas_estudio'].round().clip(0, 40).astype(int)
df_knn['Ingreso_familiar'] = df_knn['Ingreso_familiar'].round().clip(lower=0).astype(int)
df_knn['Semestres_cursados'] = df_knn['Semestres_cursados'].round().astype(int)

print("Valores faltantes después de KNN")
print(df_knn.isnull().sum())

print("\n Filas imputadas con KNN")
print(df_knn[df.isnull().any(axis=1)])
````

KNN imputa usando observaciones similares. Como algunas variables son conteos o dinero, se redondean para conservar su naturaleza.

````python
#Imputación por regresión
def imputar_por_regresion(data):
    df_reg = data.copy()
    columnas = df_reg.columns.tolist()

    for variable in df_reg.columns[df_reg.isnull().any()]:
        predictores = [col for col in columnas if col != variable]

        X = df_reg[predictores].copy()
        X = X.fillna(X.median(numeric_only=True))
        y = df_reg[variable]

        mask_missing = y.isnull()

        modelo = LinearRegression()
        modelo.fit(X.loc[~mask_missing], y.loc[~mask_missing])

        df_reg.loc[mask_missing, variable] = modelo.predict(X.loc[mask_missing])

    # Ajuste de naturaleza de los datos
    df_reg['Edad'] = df_reg['Edad'].round().astype(int)
    df_reg['Promedio'] = df_reg['Promedio'].round(2).clip(0, 5)
    df_reg['Horas_estudio'] = df_reg['Horas_estudio'].round().clip(0, 40).astype(int)
    df_reg['Ingreso_familiar'] = df_reg['Ingreso_familiar'].round().clip(lower=0).astype(int)
    df_reg['Semestres_cursados'] = df_reg['Semestres_cursados'].round().astype(int)

    return df_reg

df_regresion = imputar_por_regresion(df)

print("Valores faltantes después de regresión")
print(df_regresion.isnull().sum())

print("\n Filas imputadas con regresión")
print(df_regresion[df.isnull().any(axis=1)])
````

La regresión predice los faltantes usando las demás variables. También requiere redondeo y límites para no perder la naturaleza de cada variable.

````python
# Comparación estadística de los métodos
comparacion = pd.DataFrame({
    'Original_con_faltantes': df.mean(numeric_only=True),
    'KNN': df_knn.mean(numeric_only=True),
    'Regresion': df_regresion.mean(numeric_only=True)
}).round(2)

print(comparacion)
````

Es muy mínima la diferencia entre ambas técnicas para decidir si se preserva mejor en una o en otra la naturaleza de los datos
